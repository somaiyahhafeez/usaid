## reading csv, my data was downloaded for different fiscal years so reading them all and joining

```python
import pandas as pd

files = ["results.csv", "results (1).csv", "results (2).csv", "results (3).csv", "results (4).csv", "results (5).csv", "results (6).csv", "results (7).csv", "results (8).csv", "results (9).csv"]  # List all your CSV file names
dfs = [pd.read_csv(file) for file in files]
```

```python
full_df = pd.concat(dfs, ignore_index=True)
full_df
```

## dropping certain columns

```python
columns_to_drop = [
    'Transaction Type Name',
    'Funding Agency Acronym',
    'Managing Sub-agency or Bureau ID',
    'Implementing Partner Sub-category ID',
    'Activity End Date',
    'Income Group ID',
    'International Purpose Code',
    'Foreign Assistance Objective ID',
    'US Sector ID',
    'Implementing Partner Category ID',
    'Activity ID',
    'Managing Agency Acronym',
    'US Category ID',
    'Managing Sub-agency or Bureau Acronym',
    'International Category ID',
    'Funding Agency ID',
    'Aid Type Group ID',
    'Managing Agency ID',
    'Submission ID',
    'Country ID',
    'International Sector Code',
    'Transaction Type ID',
    'Region ID',
    'Implementing Partner ID',
    'Activity Start Date',
    'Income Group Acronym', 'Current Dollar Amount', 'Activity Project Number', 'Funding Agency Name', 'Implementing Partner Sub-category Name', 'Managing Agency Name'
]

full_df = full_df.drop(columns=columns_to_drop)

print(full_df.columns)
```

```python
funding_2024 = (
    full_df[full_df['Fiscal Year'] == 2024]
    .groupby('Country Name', as_index=False)['Constant Dollar Amount'].sum()
    .rename(columns={'Constant Dollar Amount': 'Funding 2024'})
)
funding_2015 = (
    full_df[full_df['Fiscal Year'] == 2015]
    .groupby('Country Name', as_index=False)['Constant Dollar Amount'].sum()
    .rename(columns={'Constant Dollar Amount': 'Funding 2015'})
)


funding_comparison = funding_2024.merge(funding_2015, on='Country Name', how='outer')

funding_comparison['Funding 2015 (M)'] = funding_comparison['Funding 2015'] / 1_000_000
funding_comparison['Funding 2024 (M)'] = funding_comparison['Funding 2024'] / 1_000_000

funding_comparison = funding_comparison[['Country Name', 'Funding 2015 (M)', 'Funding 2024 (M)']]

funding_comparison = funding_comparison[funding_comparison['Country Name'] != "World"]

funding_comparison['Rank 2014'] = funding_comparison['Funding 2015 (M)'].rank(ascending=False, method="min").astype('Int64')
funding_comparison['Rank 2024'] = funding_comparison['Funding 2024 (M)'].rank(ascending=False, method="min").astype('Int64')

funding_comparison = funding_comparison.nsmallest(50, 'Rank 2024')

funding_comparison.to_csv("usaid_funding_comparison.csv", index=False)
```
